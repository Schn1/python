## 傳統寫法

```python
import math
radius = eval(input("請輸入圓的半徑:")) #把輸入的字串變成數學計算型態
print("圓的面積為:", radius*radius*math.pi)
```

## 面積計算函式

```python
import math
def circle_area(radius):
    area = radius*radius*math.pi
    return area

print(circle_area(10))
```
